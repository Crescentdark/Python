```python
import pandas as pd
pd.__version__
from pandas import DataFrame
import matplotlib.pyplot as plot
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
X = DataFrame(data, columns=['production_budged_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plot.figure(figsize=(10, 6))
plot.scatter(X, y, alpha=0.3)
plot.title('Film cost vs Global revenue')
plot.xlabel('Production budgedt $')
plot.ylabel('Worldwide gross $')
plot.ylim(0, 300000000)
plot.xlim(0, 450000000)
plot.show()
```
